```python
import numpy as np
import cv2

point = ()
point1 = ()
point2 = ()
point3 = ()
point4 = ()

click_count = 0
i = 1
ids=1 

def on_mouse(event, x, y, flags, params):
	global i,point1,point2,point3,point4,point
	if event == cv2.EVENT_LBUTTONDOWN:
		point = (x,y)
		if i is 1:
			point1 = (x,y)
		elif i is 2:
			point2 = (x,y)
		elif i is 3:
			point3 = (x,y)
		elif i is 4:
			point4 = (x,y)
			i = 0
		i = i + 1



# load the image, clone it, and setup the mouse callback function
img = cv2.imread('test1.png')
cv2.namedWindow("image")
cv2.setMouseCallback("image", on_mouse)

while True:
    if len(point) != 0 :
        cv2.circle(img,point,5,(255,0,0),-1)
    if len(point4) != 0:
        vrx = np.array((point1,point2,point3,point4), np.int32)
        
        
        output = open("coordinates.txt", "a")
        
        
        output.write( "-\n          id: " + str(ids) + "\n          points: [" +
                          "[" + str(vrx[0][0]) + "," + str(vrx[0][1]) + "]," +
                          "[" + str(vrx[1][0]) + "," + str(vrx[1][1]) + "]," +
                          "[" + str(vrx[2][0]) + "," + str(vrx[2][1]) + "]," +
                          "[" + str(vrx[3][0]) + "," + str(vrx[3][1]) + "]]\n")
        output.close()
        
        vrx = vrx.reshape((-1,1,2))
        img = cv2.polylines(img, [vrx], True, (0,255,255),3)
        point1 = ()
        point2 = ()
        point3 = ()
        point4 = ()
        moments = cv2.moments(vrx)

        center = (int(moments["m10"] / moments["m00"]) - 3,
                int(moments["m01"] / moments["m00"]) + 3)

        cv2.putText(img,
                    str(ids),
                    center,
                    cv2.FONT_HERSHEY_SIMPLEX,
                    1,
                    (0,0,255),
                    2,
                    cv2.LINE_AA)
        ids+=1
        

    cv2.imshow("image", img)
    
    key = cv2.waitKey(1) & 0xFF 
    if key == ord("q"):
        cv2.destroyAllWindows()
        break
        
 
cv2.destroyAllWindows()
```
